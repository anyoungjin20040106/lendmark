```python
import pandas as pd
import geopy.distance
from sklearn.neighbors import KNeighborsClassifier
from geopy import Nominatim
import requests
import os
```

```python
tf=pd.read_csv('시장및마트현황.csv',encoding='EUC-KR')
tf
```

```python
KAKAO_API_KEY = os.getenv('KAKAO_API_KEY')

def kakao_geocode(address: str):
    url = f"https://dapi.kakao.com/v2/local/search/address.json?query={address}"
    headers = {"Authorization": f"KakaoAK {KAKAO_API_KEY}"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        result = response.json()
        if result['documents']:
            location = result['documents'][0]['address']
            return float(location['y']), float(location['x'])
    return None, None
```

```python
kakao_geocode('경기도 남양주시 와부읍 덕소리 483 외1호')
```

```python
tf['WGS84위도']=tf['소재지도로명주소'].apply(lambda x : kakao_geocode(x)[0] if x!=None else tf[tf['소재지도로명주소']==x]['WGS84위도'].values[0])
tf['WGS84경도']=tf['소재지도로명주소'].apply(lambda x : kakao_geocode(x)[1] if x!=None else tf[tf['소재지도로명주소']==x]['WGS84경도'].values[0])
tf
```

```python
tf=tf[['시장명','WGS84위도','WGS84경도']]
tf
```
